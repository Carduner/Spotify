```python
import pandas as pd
import numpy as np
```

```python
# inspect data
artist_data = pd.read_csv("data/artists.csv")
tracks_data = pd.read_csv("data/tracks.csv")
```

```python
print(artist_data.shape)
print(artist_data.info())
artist_data.head()
```

```python
print(tracks_data.shape)
print(tracks_data.info())
tracks_data.head()
```

```python
# fix list data types in both datasets

# first replace empty lists with nulls
print(artist_data.loc[artist_data.genres=='[]','genres'].shape)
artist_data.loc[artist_data.genres=='[]','genres'] = np.nan
print(artist_data.loc[artist_data.genres.isna(),'genres'].shape)
```

```python
%%time
# evaluate to an actual list
from ast import literal_eval
artist_data.loc[artist_data.genres.notna(),'genres'] = artist_data.loc[artist_data.genres.notna(),'genres'].apply(literal_eval)
for i in ['id_artists','artists']:
    tracks_data[i] = tracks_data[i].apply(literal_eval)
```

```python
# function to pull the ind position of the id
def pull_artist(x,ind):
    try:
        return x[ind]
    except:
        return np.nan
```

```python
%%time
# get first 3 artist ids from column
tracks_data.loc[:,'artist_id_1'] = tracks_data.id_artists.apply(lambda x: pull_artist(x,0))
tracks_data.loc[:,'artist_id_2'] = tracks_data.id_artists.map(lambda x: pull_artist(x,1))
tracks_data.loc[:,'artist_id_3'] = tracks_data.id_artists.map(lambda x: pull_artist(x,2))
tracks_data[[x for x in tracks_data.columns if 'artist' in x]].head()
```

### normalize data for feature store

```python
%%time
# one hot encode genres
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genres_oh = pd.DataFrame(mlb.fit_transform(artist_data.loc[artist_data.genres.notna(),'genres']),
                         columns=mlb.classes_,
                         index=artist_data.loc[artist_data.genres.notna(),'genres'].index)
genres_oh.head()
```

### apply Truncated SVD to reduce genre feature space

```python
%%time
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=100,n_iter=5,random_state=0)
svd.fit(genres_oh)
print(svd.explained_variance_ratio_.sum())
```

```python
svd.components_.shape
```

```python
svd.singular_values_.shape
```

### merge datasets together to get feature space for all songs

```python
%%time
genre_transformed = pd.DataFrame(svd.transform(genres_oh), 
                                 index = genres_oh.index)
genre_transformed.shape
```

```python
artist_data = artist_data.merge(genre_transformed,how='left',left_index=True,right_index=True)
artist_data.head()
```

```python
artist_cols = [x for x in artist_data.columns if x not in ['followers','genres','name','popularity']]
print(tracks_data.shape)
tracks_data = tracks_data.merge(artist_data[artist_cols],how='left',left_on='artist_id_1',right_on='id')
print(tracks_data.shape)
```

```python
tracks_data = tracks_data.rename(columns={'id_x':'id'})
final_cols = [x for x in tracks_data.columns if x not in ['time_signature', 'artist_id_1', 'artist_id_2', 'artist_id_3','name', 'popularity', 'duration_ms', 'explicit', 'artists',
       'id_artists', 'release_date','id_y']]
tracks_fs = tracks_data[final_cols]
tracks_fs.head()
```

```python
tracks_fs.columns.values
```

```python
tracks_fs.to_csv('data/tracks_fs_clean.csv')
```

### create recent song df in same format as above using spotipy api

```python
import spotipy
import os
%run './credentials/settings.py'
from spotipy.oauth2 import SpotifyOAuth
scope = "user-library-read"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))
def auth(scope=scope):
    SPOTIPY_CLIENT_ID = os.getenv("SPOTIPY_CLIENT_ID")
    SPOTIPY_CLIENT_SECRET = os.getenv("SPOTIPY_CLIENT_SECRET")
    SPOTIPY_REDIRECT_URI = os.getenv('SPOTIPY_REDIRECT_URI')
    sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))
    return sp
```

```python
# get liked songs
tracks_df_dict = {}
tracks_df_dict['id'] = []
tracks_df_dict['track_name'] = []
tracks_df_dict['artist_id'] = []
tracks_df_dict['artist_name'] = []
tracks_df_dict['added_at'] = []
results = sp.current_user_saved_tracks()
for idx, item in enumerate(results['items']):
    track = item['track']
    tracks_df_dict['id'].append(track['id'])
    tracks_df_dict['track_name'].append(track['name'])
    tracks_df_dict['artist_id'].append(track['artists'][0]['id'])
    tracks_df_dict['artist_name'].append(track['artists'][0]['name'])
    tracks_df_dict['added_at'].append(item['added_at'])
my_tracks = pd.DataFrame.from_dict(tracks_df_dict)
my_tracks
```

```python
# get audio features for songs
audio_features = sp.audio_features(tracks_df_dict['id'])
audio_df = pd.DataFrame(audio_features)
audio_df.head()
```

```python
audio_df.columns.values
```

```python
# merge into my_tracks
audio_cols = ['id','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
my_tracks = my_tracks.merge(audio_df[audio_cols],on='id')
my_tracks.head()
```

```python
# get genre data for artist and related artists
def related_artist_genres(df,scope=scope):
    sp = auth()
    results = {} # initiate container for results
    for artist_id in df.artist_id:
        genre_set = set() # use set so we don't dupe genres
        try:
            artist = sp.artist(artist_id) # get artist genre
            genre_set.update(artist['genres'])
        except:
            continue
        try:
            df_related_artists = pd.DataFrame(sp.artist_related_artists(artist_id)['artists']) # get related artist
            # update set with all genres from related artists
            for i in df_related_artists.genres:
                genre_set.update(i)
        except:
            continue
        results[artist_id] = list(genre_set)
    return results
```

```python
%%time
genre_results = related_artist_genres(my_tracks)
```

```python
# create flattened df
my_genres = pd.Series(genre_results).to_frame('genres')
my_genres.head()
```

```python
%%time
# transform genres with mlb.transform that was fitted early to keep things consistent
my_genres_oh = pd.DataFrame(mlb.transform(my_genres.genres),
                         columns=mlb.classes_,
                         index=my_genres.index)
my_genres_oh.head()
```

```python
%%time
# reduce genres with svd
my_genres_transformed = pd.DataFrame(svd.transform(my_genres_oh), 
                                 index = my_genres_oh.index)
my_genres_transformed = my_genres_transformed.reset_index()
my_genres_transformed = my_genres_transformed.rename(columns={'index':'artist_id'})
print(my_genres_transformed.shape)
my_genres_transformed.head()
```

```python
# merge all together
print(my_tracks.shape)
my_tracks = my_tracks.merge(my_genres_transformed,on='artist_id')
print(my_tracks.shape)
```
