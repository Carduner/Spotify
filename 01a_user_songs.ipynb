### Recent User Songs and Create Playlist

```python
import spotipy
import pandas as pd
import os
%run './credentials/settings.py'
from spotipy.oauth2 import SpotifyOAuth
scope = "user-library-read"
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=os.getenv("SPOTIPY_CLIENT_ID"),
                                               client_secret=os.getenv("SPOTIPY_CLIENT_SECRET"),
                                               redirect_uri=os.getenv('SPOTIPY_REDIRECT_URI_READ'),
                                               scope=scope))
```

```python
# get liked songs
results = sp.current_user_saved_tracks()
tracks_df_dict = {}
tracks_df_dict['id'] = []
tracks_df_dict['track_name'] = []
tracks_df_dict['artist_id'] = []
tracks_df_dict['artist_name'] = []
tracks_df_dict['added_at'] = []
results = sp.current_user_saved_tracks()
for idx, item in enumerate(results['items']):
    track = item['track']
    tracks_df_dict['id'].append(track['id'])
    tracks_df_dict['track_name'].append(track['name'])
    tracks_df_dict['artist_id'].append(track['artists'][0]['id'])
    tracks_df_dict['artist_name'].append(track['artists'][0]['name'])
    tracks_df_dict['added_at'].append(item['added_at'])
my_tracks = pd.DataFrame.from_dict(tracks_df_dict)
my_tracks
```

```python
# get audio features for songs
audio_features = sp.audio_features(tracks_df_dict['id'])
audio_df = pd.DataFrame(audio_features)
audio_df.head()
```

```python
# merge into my_tracks
audio_cols = ['id','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
my_tracks = my_tracks.merge(audio_df[audio_cols],on='id')
my_tracks.head()
```

```python
# get genre data for artist and related artists
def related_artist_genres(df,sp):
    results = {} # initiate container for results
    for artist_id in df.artist_id:
        genre_set = set() # use set so we don't dupe genres
        try:
            artist = sp.artist(artist_id) # get artist genre
            genre_set.update(artist['genres'])
        except:
            continue
        try:
            df_related_artists = pd.DataFrame(sp.artist_related_artists(artist_id)['artists']) # get related artist
            # update set with all genres from related artists
            for i in df_related_artists.genres:
                genre_set.update(i)
        except:
            continue
        results[artist_id] = list(genre_set)
    return results
```

```python
%%time
genre_results = related_artist_genres(my_tracks,sp)
```

```python
# create flattened df
my_genres = pd.Series(genre_results).to_frame('genres')
my_genres.head()
```

```python
# import transformers
import pickle
transformers = pickle.load(open('objects/objects_20210709131218.pkl','rb'))
transformers.keys()
```

```python
mlb = transformers['mlb']
svd = transformers['svd']
```

```python
%%time
# transform genres with mlb.transform that was fitted early to keep things consistent
my_genres_oh = pd.DataFrame(mlb.transform(my_genres.genres),
                         columns=mlb.classes_,
                         index=my_genres.index)
my_genres_oh.head()
```

```python
%%time
# reduce genres with svd
my_genres_transformed = pd.DataFrame(svd.transform(my_genres_oh), 
                                 index = my_genres_oh.index)
my_genres_transformed = my_genres_transformed.reset_index()
my_genres_transformed = my_genres_transformed.rename(columns={'index':'artist_id'})
print(my_genres_transformed.shape)
my_genres_transformed.head()
```

```python
# merge all together
print(my_tracks.shape)
my_tracks = my_tracks.merge(my_genres_transformed,on='artist_id')
print(my_tracks.shape)
```

### calculate similarity with recent songs and all songs

```python
tracks_fs = pd.read_csv('data/tracks_fs_clean.csv',index_col=0)
tracks_fs.head()
```

```python
# check for nulls
for i in tracks_fs.columns:
    print(i, tracks_fs[i].isna().sum())
```

```python
# check for nulls
for i in my_tracks.columns:
    print(i, my_tracks[i].isna().sum())
```

```python
# only nulls in tracks_fs are for genres. will just fill those with null as it means it didn't have a genre
tracks_fs = tracks_fs.fillna(0)
```

```python
final_cols = [x for x in tracks_fs.columns if 'id' not in x]
```

```python
# replace cols with string of 1-100
replace = {}
for i in my_tracks.columns:
    if i in range(0,100):
        replace[i] = str(i)
my_tracks = my_tracks.rename(columns=replace)
```

```python
# just find songs similar to last song liked (stay)
Y = my_tracks[final_cols].head(1)
Y
```

```python
%%time
# just find songs similar to last song liked (stay)
from sklearn.metrics.pairwise import cosine_similarity
X = tracks_fs[final_cols]
sim_matrix = cosine_similarity(X,Y)
```

```python
sim_matrix.shape
```

```python
df_sim = pd.DataFrame(sim_matrix, index=tracks_fs.index)
df_sim['sum'] = df_sim.sum(axis=1)
df_sim.head()
```

```python
tracks_data = pd.read_csv('data/tracks.csv')
print(tracks_data.shape)
tracks_data.head()
```

```python
df_sim['id'] = tracks_data['id']
df_sim['artists'] = tracks_data['artists']
df_sim['name'] = tracks_data['name']
```

```python
rec = df_sim[['id','name','artists','sum']].sort_values(by=['sum'],ascending=False).head(50)
rec
```

```python
# get popularity
tracks = sp.tracks(rec.head(50).id.values.tolist())
```

```python
popularity = {}
for i in range(len(tracks['tracks'])):
    popularity[tracks['tracks'][i]['id']] = tracks['tracks'][i]['popularity']
```

```python
df_pop = pd.DataFrame.from_dict(popularity,orient='index').reset_index()
df_pop.columns = ['id','popularity']
rec = rec.merge(df_pop,on='id')
rec.head()
```

```python
rec_de_dupe = rec.sort_values(by='popularity',ascending=False).groupby(['artists','name']).first()
print(rec_de_dupe.shape)
rec_de_dupe.head()
```

```python
final_recs = rec_de_dupe.reset_index().sort_values(by='sum',ascending=False)
final_recs
```

```python
final_recs.to_csv('data/recommended_playlist_0712.csv',index=False)
```
